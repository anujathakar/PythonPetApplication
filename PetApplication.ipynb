```python
Pets Inventory Application

1.	Add a new pet
2.	Compute and print current weight of all pets in the system
3.	Print all pet information for every pet in the system  
4.  Print owner information and pet names for all owners with more than one pet.
5.	Exit the application 
The first time the program is executed, file will have no pets in it. Second time onwards, all pet data from previous executions of the program, should be available

Pet Application
A base class `Pet` has the following attributes:
petName (String)
date of birth (Date). The user will enter year, month and date. 
birth weight (float). Note that the user will enter this in ounces.
owner (Owner).
The child classes `Mammal`, `Fish` and `Amphibian` inherit from `Pet` have the following additional properties 
Mammal –
. litter_size (int),
. hasClaws (bool),

Fish -
. scale_condition (one of 2 values: smooth, rough),
. length (float)

Amphibian –
. number of limbs (int),
. isVenomous (bool)

All these classes have the appropriate __init_\() method. In addition, they have additional methods required to complete the application as described below.

First find the number of days since birth for each pet type. Then compute the current weight as follows:

Mammal: Find the number of days since birth. Weight increases by 10% every 60 days for the first 300 days. Weight is constant after that. Assume that the increase occurs on each 60th day only.

Fish: Weight increases by 5% every 80 days for the first 240 days and then stays constant. Assume that the increase occurs on each 80th day only.

Amphibian: Weight increases by 5% every 120 days for the first 360 days and then increases by 3% every 120 days. Assume that the increase occurs on each 120th day only.

TheOwner class is described next. The Owner class has two instance variables – person name and address.  
__Once the fives classes, Pet, Mammal, Fish, Amphibian and Person have all been defined and tested per the requirements stated above, write a driver program to do the following:
1. Add a new pet
2. Print pet name, date of birth and owner name for all pets in the system. In addition, litter size and whether or not the pet has claws should be displayed for Mammals.The length of the Fish should be displayed for Fish type and number of limbs should be displayed for Amphibians.
3. Compute and print current weight (in pounds) of all pets in the system.
4. Print owner information and pet names for all owners with more than one pet.
5. Exit the application after writing all information to a file

The program must repeatedly display the above options and allow the user to select one of them until she chooses option 5. The program must save all data to a file and then exit the system once the user selects option 5.

Note: Any pet data from earlier executions of the application must be available in subsequent executions.


The program uses exception handling to validate user input for the following variables:

Date of birth: Valid Date
Birth Weight: Positive number
Litter Size: Positive Integer
hasClaws: User should enter only 'y' or 'n' in either upper or lower case.
scaleCondition: User should only enter 's' or 'r' in either upper or lower case
length: Positive integer
number of limbs: Positive integer
isVenomous: User should enter only 'y' or 'n' in either upper or lower case.
If the user enters invalid values for any of the variables above, print an appropriate message and ask the user to re-enter a valid value.
The program also ensures that the user enters valid values for the menu options as well as the choice of pet type. If the user enters incorrect options for either the menu option or the pet type,the program displays an appropriate message and display the menu or the choice of pet types again.
```

```python
from datetime import date
import pickle

class Owner:
    def __init__(self, person_name, address):
        self.person_name = person_name
        self.address = address

class Pet:
    def __init__(self, name_pet, dob_year, dob_month, dob_day, birth_weight_ounces, person_name, person_address):
        self.name_pet = name_pet
        self.date_of_birth = date(dob_year, dob_month, dob_day)
        self.birth_weight_ounces = birth_weight_ounces
        self.owner = Owner(person_name, person_address)
        self.age = self.days_since_birth_compute()

    def days_since_birth_compute(self):
        today = date.today()
        days_since_birth = (today - self.date_of_birth).days
        return days_since_birth

    def current_weight_compute(self):
        return self.birth_weight_ounces / 16.0  # Convert ounces to pounds

class Mammal(Pet):
    def __init__(self, name_pet, dob_year, dob_month, dob_day, birth_weight_ounces, person_name, person_address, litter_size, has_claws):
        super().__init__(name_pet, dob_year, dob_month, dob_day, birth_weight_ounces, person_name, person_address)
        self.litter_size = litter_size
        self.has_claws = has_claws

    def current_weight_compute(self):
        days_since_birth = self.days_since_birth_compute()

        # Weight increases by 10% every 60 days for the first 300 days
        if days_since_birth <= 300:
            cycles = days_since_birth // 60
            increase = cycles * 0.10
            new_weight = self.birth_weight_ounces * (1 + increase)
        else:
            # Weight is constant after 300 days
            new_weight = self.birth_weight_ounces

        return new_weight / 16.0  # Convert ounces to pounds

class Fish(Pet):
    def __init__(self, name_pet, dob_year, dob_month, dob_day, birth_weight_ounces, person_name, person_address, scale_condition, length):
        super().__init__(name_pet, dob_year, dob_month, dob_day, birth_weight_ounces, person_name, person_address)
        self.scale_condition = scale_condition
        self.length = length

    def current_weight_compute(self):
        days_since_birth = self.days_since_birth_compute()

        # Weight increases by 5% every 80 days for the first 240 days
        if days_since_birth <= 240:
            cycles = days_since_birth // 80
            increase = cycles * 0.05
            new_weight = self.birth_weight_ounces * (1 + increase)
        else:
            # Weight is constant after 240 days
            new_weight = self.birth_weight_ounces

        return new_weight / 16.0  # Convert ounces to pounds

class Amphibian(Pet):
    def __init__(self, name_pet, dob_year, dob_month, dob_day, birth_weight_ounces, person_name, person_address, num_limbs, is_venomous):
        super().__init__(name_pet, dob_year, dob_month, dob_day, birth_weight_ounces, person_name, person_address)
        self.num_limbs = num_limbs
        self.is_venomous = is_venomous

    def current_weight_compute(self):
        days_since_birth = self.days_since_birth_compute()

        # Weight increases by 5% every 120 days for the first 360 days
        if days_since_birth <= 360:
            cycles = days_since_birth // 120
            increase = cycles * 0.05
            new_weight = self.birth_weight_ounces * (1 + increase)
        else:
            # Weight increases by 3% every 120 days after 360 days
            cycles_after_360 = (days_since_birth - 360) // 120
            increase_after_360 = cycles_after_360 * 0.03
            new_weight = self.birth_weight_ounces * (1 + 0.05 + increase_after_360)

        return new_weight / 16.0  # Convert ounces to pounds

def save_data_to_file(pets):
    with open("pet_data.dat", "wb") as file:
        pickle.dump(pets, file)

def load_data_from_file():
    try:
        with open("pet_data.dat", "rb") as file:
            pets = pickle.load(file)
    except FileNotFoundError:
        pets = []
    return pets


def main():
    pets = load_data_from_file()

    while True:
        print("\nOptions:")
        print("1. To Add a Pet")
        print("2. To print pet information of all pet in the database")
        print("3. To print pet name and current weight of all the pets in the databasw")
        print("4. To print pet name, owner name and owner address for all owners with more than one pet")
        print("5. To exit the program")

        choice = input("Your Selection: ")

        if choice == '1':
            # Add a new pet
            type_pet = input("Enter pet type (Mammal, Fish, Amphibian): ")
            name_pet = input("Enter pet name: ")
            dob_year = int(input("Enter birth year: "))
            dob_month = int(input("Enter birth month: "))
            dob_day = int(input("Enter birth day: "))
            birth_weight = float(input("Enter birth weight in ounces: "))
            person_name = input("Enter owner name: ")
            person_address = input("Enter owner address: ")

            try:
                if type_pet.lower() == 'mammal':
                    litter_size = int(input("Enter litter size: "))
                    has_claws = input("Does the pet have claws? (yes/no): ").lower()
                    if has_claws not in ['yes', 'no']:
                        raise ValueError("Please enter 'yes' or 'no'.")
                    new_pet = Mammal(name_pet, dob_year, dob_month, dob_day, birth_weight, person_name, person_address, litter_size, has_claws == 'y')
                elif type_pet.lower() == 'fish':
                    scale_condition = input("Enter scale condition (smooth/rough): ").lower()
                    if scale_condition not in ['smooth', 'rough']:
                        raise ValueError("Please enter 'smooth' or 'rough'.")
                    length = float(input("Enter length: "))
                    new_pet = Fish(name_pet, dob_year, dob_month, dob_day, birth_weight, person_name, person_address, scale_condition, length)
                elif type_pet.lower() == 'amphibian':
                    num_limbs = int(input("Enter number of limbs: "))
                    is_venomous = input("Is the pet venomous? (yes/no): ").lower()
                    if is_venomous not in ['yes', 'no']:
                        raise ValueError("Please enter 'yes' or 'no'.")
                    new_pet = Amphibian(name_pet, dob_year, dob_month, dob_day, birth_weight, person_name, person_address, num_limbs, is_venomous == 'y')
                else:
                    raise ValueError("Invalid pet type. Please enter Mammal, Fish, or Amphibian.")

                pets.append(new_pet)
                print("New pet added!")

            except ValueError as e:
                print(f"Error: {e}")

        elif choice == '2':
            # Print pet information
            print("\nPet Information:")

            mammal_header = False
            fish_header = False
            amphibian_header = False

            for pet in pets:
                if isinstance(pet, Mammal):
                    if not mammal_header:
                        print("------------------------------------------------------------------------------------------------------------")
                        print("Data for all Mammals in the database")
                        print("------------------------------------------------------------------------------------------------------------")
                        mammal_header = True
                    print(f"\nPet Name: {pet.name_pet} \tDate of Birth: {pet.date_of_birth} \tBirth Weight: {pet.birth_weight_ounces} ounces \tOwner: {pet.owner.person_name} \tOwner Address: {pet.owner.address} \tLitter Size: {pet.litter_size} \tHas Claws: {'Yes' if pet.has_claws else 'No'}")
                elif isinstance(pet, Fish):
                    if not fish_header:
                        print("-------------------------------------------------------------------------------------------------------------")
                        print("Data for all Fish in the database")
                        print("-------------------------------------------------------------------------------------------------------------")
                        fish_header = True
                    print(f"\nPet Name: {pet.name_pet} \tDate of Birth: {pet.date_of_birth} \tBirth Weight: {pet.birth_weight_ounces} ounces \tOwner: {pet.owner.person_name} \tOwner Address: {pet.owner.address} \tScale Condition: {pet.scale_condition} \tLength: {pet.length} inches")
                elif isinstance(pet, Amphibian):
                    if not amphibian_header:
                        print("--------------------------------------------------------------------------------------------------------------")
                        print("Data for all Amphibians in the database")
                        print("---------------------------------------------------------------------------------------------------------------")
                        amphibian_header = True
                        print(f"\nPet Name: {pet.name_pet} \tDate of Birth: {pet.date_of_birth} \tBirth Weight: {pet.birth_weight_ounces} ounces \tOwner: {pet.owner.person_name} \tOwner Address: {pet.owner.address} \tNumber of Limbs: {pet.num_limbs} \tIs Venomous: {'Yes' if pet.is_venomous else 'No'}")

        elif choice == '3':
            # Compute and print current weight of all pets
            print("\nCurrent Weight of All Pets (in pounds):")
            for pet in pets:
                current_weight = pet.current_weight_compute()
                print(f"Pet Name: {pet.name_pet}, Current Weight: {current_weight:.2f} lbs")

        elif choice == '4':
            # Print owner information for owners with more than one pet
            owner_pets_count = {}
            for pet in pets:
                owner_key = (pet.owner.person_name, pet.owner.address)
                owner_pets_count[owner_key] = owner_pets_count.get(owner_key, 0) + 1

            print("-----------------------------------------------------------------------------------------------------")
            print("\nPet name-Owner Name-Owner Address for Multi Pet Owners")
            print("-----------------------------------------------------------------------------------------------------")
            for owner_key, count in owner_pets_count.items():
                if count > 1:
                    person_name, person_address = owner_key
                    print(f"{person_name} lives in {person_address} and has {count} pets. The pet names are:")
                    for pet in pets:
                        if (pet.owner.person_name, pet.owner.address) == owner_key:
                            print(f"  {pet.name_pet}")

        elif choice == '5':
            # Save data to a .dat file and exit
            save_data_to_file(pets)
            print("Exiting the application. Data saved to 'pet_data.dat'.")
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 5.")
```
